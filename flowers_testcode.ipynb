```python
# 1. Import Libraries
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image

# 2. Load Model .h5
model_path = '/content/best_model_flowers.h5'        # ganti sesuai lokasi file .h5 kamu
model = load_model(model_path)

# 3. Set Kelas (Sesuai urutan training)
class_names = ['daisy', 'dandelion', 'rose', 'sunflower', 'tulip']

# 4. Path Gambar yang ingin diuji
img_path = '/content/drive/MyDrive/Artificial Intelligence/dataset/flowers/daisy/506348009_9ecff8b6ef.jpg'  # contoh: gambar daisy

# 5. Preprocess Gambar
img_size = (128, 128)  # Sesuaikan jika model dibuat dengan size lain
img = image.load_img(img_path, target_size=img_size)
img_array = image.img_to_array(img) / 255.0
img_array = np.expand_dims(img_array, axis=0)

# 6. Prediksi
pred = model.predict(img_array)
pred_class_idx = np.argmax(pred)
pred_class = class_names[pred_class_idx]
prob = pred[0][pred_class_idx]

# 7. Tampilkan Hasil
plt.imshow(image.array_to_img(img))
plt.title(f"Predicted: {pred_class} ({prob:.2%})")
plt.axis('off')
plt.show()

print(f'Predicted class: {pred_class} dengan probabilitas {prob:.2%}')
```
